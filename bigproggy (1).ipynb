```python
#Program to random colldie particle together
import numpy as np

# Function for collision A+B=2C
def A(matrix):
    if matrix[0, 0] == 0 or matrix[1, 0] == 0:
        return matrix, "Collision A+B is not possible - Cannot perform when A=0 or B=0"
    else:
        matrix[0, 0] -= 1
        matrix[1, 0] -= 1
        matrix[2, 0] += 2
        return matrix, "Collision A=B=2C"

# Function for collision A+C=2B
def B(matrix):
    if matrix[0, 0] == 0 or matrix[2, 0] == 0:
        return matrix, "Collision A+C=2B is not possible- Cannot perform when A=0 or C=0"
    else:
        matrix[0, 0] -= 1
        matrix[1, 0] += 2
        matrix[2, 0] -= 1
        return matrix, "collision A+C=2B."

# Function for collision B+C=2A
def C(matrix):
    if matrix[1, 0] == 0 or matrix[2, 0] == 0:
        return matrix, "Collsiosn B+C=2A is not possible - Cannot perform when B=0 or C=0"
    else:
        matrix[0, 0] += 2
        matrix[1, 0] -= 1
        matrix[2, 0] -= 1
        return matrix, "Collision B+C=2A"

collis = [A, B, C]
num_iter = 10

print("Enter the starting values for the matrix (3 comma-separated integers):")
start_val = list(map(int, input().split(',')))
matrix = np.array(start_val).reshape(3, 1)

for i in range(num_iter):
    # Check for finial case
    if np.sum(matrix == 0) >= 2:
        print("Finial case achieved.")
        break
    
    # Each collision has equal probability
    chosen_collis = np.random.choice(collis)
    matrix, collis_des = chosen_collis(matrix)
    print(f"Iteration {i+1}: {collis_des}")
    print("Matrix after action:")
    print(matrix)
    print()
```

```python
#Program to inspect every collision possible 
import numpy as np
from itertools import product

# Function for collision A+B=2C
def A(matrix):
    if matrix[0, 0] == 0 or matrix[1, 0] == 0:
        return False, matrix, "Collision A+B is not possible - Cannot perform when A=0 or B=0"
    else:
        matrix[0, 0] -= 1
        matrix[1, 0] -= 1
        matrix[2, 0] += 2
        return True, matrix, "Collision A+B=2C"

#Function for collision A+C=2B
def B(matrix):
    if matrix[0, 0] == 0 or matrix[2, 0] == 0:
        return False, matrix, "Collision A+C=2B is not possible- Cannot perform when A=0 or C=0"
    else:
        matrix[0, 0] -= 1
        matrix[1, 0] += 2
        matrix[2, 0] -= 1
        return True, matrix, "collision A+C=2B."

#  Function for collision B+C=2A
def C(matrix):
    if matrix[1, 0] == 0 or matrix[2, 0] == 0:
        return False, matrix, "Collsiosn B+C=2A is not possible - Cannot perform when B=0 or C=0"
    else:
        matrix[0, 0] += 2
        matrix[1, 0] -= 1
        matrix[2, 0] -= 1
        return True, matrix, "Collision B+C=2A"

# all possible collisions, length 2
collis_combo = list(product(['A', 'B', 'C'], repeat=3))

num_iter = 10

# Ask user for starting matrix
print("Enter the starting values for the matrix (3 comma-separated integers):")
start_val = list(map(int, input().split(',')))
initial_matrix = np.array(start_val).reshape(3, 1)

for combination in collis_combo:
    matrix = initial_matrix.copy()
    print(f"Initial matrix: {matrix.flatten()}")
    print(f"Applying actions: {' -> '.join(combination)}")
    for collis_name in combination:
        if collis_name == 'A':
            can_apply, matrix, collis_des = A(matrix)
        elif collis_name == 'B':
            can_apply, matrix, collis_des = B(matrix)
        elif collis_name == 'C':
            can_apply, matrix, collis_des = C(matrix)
        if not can_apply:
            print(f"Invalid action sequence. {collis_des}")
            break
        print(collis_des)
        print("Matrix after action:")
        print(matrix)
        print()
    print()
```
